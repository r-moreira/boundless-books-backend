```python
import json
from ebooklib import epub

def extract_epub_metadata_and_cover_image(file_path):
    book = epub.read_epub(file_path)
    
    metadata_fields = [
        'title', 'language', 'creator', 'publisher', 'date', 'description'
    ]
    
    metadata = {}
    for field in metadata_fields:
        metadata[field] = book.get_metadata('DC', field)
    

    cover_images = list(book.get_items_of_type(ebooklib.ITEM_IMAGE))
    
    if cover_images:
        cover_images = cover_images[0].get_content()
        cover_image_path = f'{file_path}.cover.jpg'
        with open(cover_image_path, 'wb') as f:
            f.write(cover_images)
            print(f'Cover image saved at {cover_image_path}')
        
    metadata = {
        "title": metadata["title"][0][0] if metadata["title"] else None,
        "language": metadata["language"][0][0] if metadata["language"] else None,
        "creator": metadata["creator"][0][0] if metadata["creator"] else None,
        "publisher": metadata["publisher"][0][0] if metadata["publisher"] else None,
        "date": metadata["date"][0][0] if metadata["date"] else None,
        "description": metadata["description"][0][0] if metadata["description"] else None
    }

    return json.dumps(metadata, indent=2, ensure_ascii=False)

```

```python
import epub_metadata
import base64
from PIL import Image
from io import BytesIO

epub = epub_metadata.epub('/Users/rodrigo/Dev/infnet/extension/boundless-books/notebook/Animais Fantasticos & Onde Habitam - J.K. Rowling.EPUB')

cover_base64 = epub.metadata.cover

cover_data = base64.b64decode(cover_base64)

cover_image = Image.open(BytesIO(cover_data))

cover_image_path = f"{epub.metadata.title} - {epub.metadata.creator}_cover.jpeg"
cover_image.save(cover_image_path)

print(f"Cover image saved at {cover_image_path}")
```
